```python
#Project #2: Automation of Concrete Mix Design.
#Client: Nebraska Department of Transportation (NDOT).
#All calculations are performed for 1 cubic yard of concrete 

#Project identification
project_name = "Project #2: Automation of Concrete Mix Design"
client = "Nebraska Department of Transportation (NDOT)"

#Display project information 
print(f"Project Name: {project_name}")
print(f"Client: {client}")

#Fundamental Engineering Constants
#Total volume of 1 cubic yard of concrete (ft³)
cubic_yard_ft3 = 27

#Unit weight of water (lb/ft³)
unit_weight_water = 62.4

#Default specific gravities for SCM's
sg_fly_ash_K = 2.30
sg_silica_L = 2.20
sg_SCM_M = 2.60

#Display constants
print("\nDefined Engineering Constants:")
print(f" - 1 cubic yard = {cubic_yard_ft3} ft³")
print(f" - Unit weight of water = {unit_weight_water} lb/ft³")
print(f" - Default SG (Fly Ash) = {sg_fly_ash_K}")
print(f" - Default SG (Silica Fume) = {sg_silica_L}")
print(f" - Default SG (Other SCM) = {sg_SCM_M}")
```

```python
#Engineering Functions.
#Each of the functions replaces the specific Excel formula from the NDOT Mix Design worksheet.
#Each function corresponds to the specific calculation for 1 cubic yard of concrete.

#Water Weight (Q)
#Q = (A + B + C + D) * E
def calculate_water_weight_Q(cement_A, fly_ash_B, silica_C, other_scm_D, wc_ratio_E):
    water_weight_Q = (cement_A + fly_ash_B + silica_C + other_scm_D) * wc_ratio_E
    return water_weight_Q
    
#Cement Volume (R)
#R = A / (J * 62.4)
def calculate_volume_R(cement_A, sg_cement_J):
    volume_cement_R = cement_A / (sg_cement_J * unit_weight_water)
    return volume_cement_R

#Fly Ash Volume (S)
#S = B / (K * 62.4)
def calculate_volume_S(fly_ash_B, sg_fly_ash_K):
    volume_fly_ash_S = fly_ash_B / (sg_fly_ash_K * unit_weight_water)
    return volume_fly_ash_S
    
#Silica Fume Volume (T)
#T = C / (L * 62.4)
def calculate_silica_volume_T(silica_C, sg_silica_L):
    volume_silica_fume_T = silica_C / (sg_silica_L * unit_weight_water)
    return volume_silica_fume_T

#Other SCM Volume (U)
#U = D / (M * 62.4)
def calculate_scm_volume_U(other_scm_D):
    volume_scm_U = other_scm_D / (sg_SCM_M * unit_weight_water)
    return volume_scm_U

#Air Volume (V)
#V = (F / 100) * 27
def air_volume_V(air_content_F):
    volume_V = (air_content_F / 100) * cubic_yard_ft3
    return volume_V

#Water Volume (W)
#W = Q / 62.4
def water_volume_w(weight_of_water_Q):
    volume_W = weight_of_water_Q / unit_weight_water
    return volume_W

#Total Aggregate Volume (X)
#X = 27 - R - S - T - U - V - W
def aggregate_volume_X(volume_R, volume_S, volume_T, volume_U, volume_V, volume_W):
    volume_X = cubic_yard_ft3 - volume_R - volume_S - volume_T - volume_U - volume_V - volume_W
    return volume_X

#Fine Aggregate Weight (Y)
#Y = 62.4 * (G/100) * N * X
def calculate_fine_aggregate_weight_Y(percent_fine_G, sg_fine_N, total_agg_volume_X):
    fraction_fine = percent_fine_G / 100
    weight_Y = unit_weight_water * fraction_fine * sg_fine_N * total_agg_volume_X
    return weight_Y

#Coarse Aggregate Weight (Z)
#Z = 62.4 * (H/100) * O * X
def calculate_coarse_aggregate_weight_Z(percent_coarse_H, sg_coarse_O, total_agg_volume_X):
    fraction_coarse = percent_coarse_H / 100
    weight_Z = unit_weight_water * fraction_coarse * sg_coarse_O * total_agg_volume_X
    return weight_Z

#Other Aggregate Weight (AA)
#AA = 62.4 * (I/100) * P * X
def calculate_other_aggregate_weight_AA(percent_other_I, sg_other_P, total_agg_volume_X):
    fraction_other = percent_other_I / 100
    weight_AA = unit_weight_water * fraction_other * sg_other_P * total_agg_volume_X
    return weight_AA
```

```python
#This section gathers all required mix design inputs from the user.
#Each input is converted to a numeric value so it can be used in the engineering calculations that follow

print("\n--- NDOT Mix Design Input Collection ---")

#Project information
project_no = int(input("Enter Project Number: "))
concrete_class = input("Enter Class of Concrete: ")

#Cementitious material weights (lb per cubic yard)
cement_A = float(input("Enter Cement Weight A (lb/yd³): "))
fly_ash_B = float(input("Enter Fly Ash Weight B (lb/yd³): "))
silica_C = float(input("Enter Silica Fume Weight C (lb/yd³): "))
other_scm_D = float(input("Enter Other SCM Weight D (lb/yd³): "))

#Design parameters
wc_ratio_E = float(input("Enter Water–Cement Ratio E: "))
air_content_F = float(input("Enter Air Content F (%): "))

#Aggregate proportions (%)
percent_fine_G = float(input("Enter Percent Fine Aggregate G (%): "))
percent_coarse_H = float(input("Enter Percent Coarse Aggregate H (%): "))
percent_other_I = float(input("Enter Percent Other Aggregate I (%): "))

#Specific gravities for cementitious materials
sg_cement_J = float(input("Enter Specific Gravity of Cement J: "))
sg_fly_ash_K = float(input("Enter Specific Gravity of Fly Ash K: "))
sg_silica_L = float(input("Enter Specific Gravity of Silica Fume L: "))
sg_SCM_M = float(input("Enter Specific Gravity of Other SCM M: "))

#Specific gravities for aggregates
sg_fine_N = float(input("Enter Specific Gravity of Fine Aggregate N: "))
sg_coarse_O = float(input("Enter Specific Gravity of Coarse Aggregate O: "))
sg_other_P = float(input("Enter Specific Gravity of Other Aggregate P: "))

print("\nAll inputs successfully collected.\n")
```

```python
#This section completes all the engineering calculations using the user inputs collected prior, as well as the functions defined earlier.
#The workflow follows the exact sequence of the NDOT Mix Design Excel worksheet.

print("\n--- Running NDOT Mix Design Calculations ---")

#Step 1: Water Weight (Q)
Q = calculate_water_weight_Q(cement_A, fly_ash_B, silica_C, other_scm_D, wc_ratio_E)

#Step 2: Cementitious Material Volumes (R, S, T, U)
R = calculate_volume_R(cement_A, sg_cement_J)
S = calculate_volume_S(fly_ash_B, sg_fly_ash_K)
T = calculate_silica_volume_T(silica_C, sg_silica_L)
U = calculate_scm_volume_U(other_scm_D)

#Step 3: Air Volume (V) and Water Volume (W)
V = air_volume_V(air_content_F)
W = water_volume_w(Q)

#Step 4: Total Aggregate Volume (X)
X = aggregate_volume_X(R, S, T, U, V, W)

#Step 5: Aggregate Weights (Y, Z, AA)
Y = calculate_fine_aggregate_weight_Y(percent_fine_G, sg_fine_N, X)
Z = calculate_coarse_aggregate_weight_Z(percent_coarse_H, sg_coarse_O, X)
AA = calculate_other_aggregate_weight_AA(percent_other_I, sg_other_P, X)

print("\nAll calculations completed successfully.\n")



```
